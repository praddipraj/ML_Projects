```python
sex=pd.get_dummies(data['Sex'],drop_first=True)
sex = sex.astype(int)

pclass=pd.get_dummies(data['Pclass'],drop_first=True)
pclass = pclass.astype(int)

Embark=pd.get_dummies(data['Embarked'],drop_first=True)
Embark = Embark.astype(int)


data=pd.concat([data,sex,pclass,Embark],axis=1)
data
```

```python
data.drop(['Name','PassengerId','Pclass','Sex','Embarked','Ticket'],axis=1,inplace=True)
print(data)
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(-1,inplace=True)
data.isnull().sum()
```

```python
data=data.drop('Cabin',axis=1)
```

```python
data
```

```python
x=data.drop('Survived',axis=1)
y=data['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train.columns = x_train.columns.astype(str)
```

```python
x_train.columns = x_train.columns.astype(str)
x_test.columns = x_test.columns.astype(str)
model=LogisticRegression(max_iter=100000)
model.fit(x_train,y_train)
prediction=model.predict(x_test)
classification_report(y_test,prediction)
print("Test Accuracy",accuracy_score(y_test,prediction))
print("Test Precision",precision_score(y_test,prediction))
print("Test Recall : ",recall_score(y_test,prediction))
print("Test F1 score : ",f1_score(y_test,prediction))
```

```python
confusion_matrix(y_test,prediction)
```

```python
predictions=model.predict(x_train)
classification_report(y_train,predictions)
print("Test Accuracy",accuracy_score(y_train,predictions))
print("Test Precision",precision_score(y_train,predictions))
print("Test Recall : ",recall_score(y_train,predictions))
print("Test F1 score : ",f1_score(y_train,predictions))
```
